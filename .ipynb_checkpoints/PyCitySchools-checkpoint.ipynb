# Analysis
    -School size has a large impact on math and reading scores.  Only 58% of students at Large schools (more than 5000 students) passed both reading and math, compared to roughly 90% of small and medium size schools
    
    -Interestingly, having a higher budget per student does not increase the likelihood that they will pass reading and math.  Only 54% of students at schools that spend more than 645 dollars passed both math and reading, compared to 90% of students at schools that spend less than 585.  However, given that the buckets for spend are group closely together, I would use caution coming to the conclusion that investing in students does not yield results.

```python
#Pull in dependencies and set path for reading files
import pandas as pd
school_path = 'Resources/schools_complete.csv'
student_path = 'Resources/students_complete.csv'
```

```python
#read in csv file and store in dataframe
student_data = pd.read_csv(student_path)
school_data = pd.read_csv(school_path)
school_data.head()
```

```python
#merge the two csv files into a single dataframe
school_complete= pd.merge(student_data,school_data,how='left',on=['school_name','school_name'])
school_complete.head()
```

## District Summary

```python
#find the number of schools
school_count = len(school_complete['school_name'].unique())
school_count
```

```python
#find the number of students
total_students = school_complete['Student ID'].count()
total_students
```

```python
#find the total budget across all schools
total_budget = school_data['budget'].sum()
total_budget
```

```python
#find the average_math_score
average_math_score = school_complete['math_score'].mean()
average_math_score
```

```python
#find the average reading score
average_reading_score = school_complete['reading_score'].mean()
average_reading_score
```

```python
#create variable that filters for passing math scores - greater than 70
passing_math = school_complete.loc[(school_complete['math_score'] >= 70),:]
percent_passing_math = (passing_math['Student ID'].count() / total_students) *100
percent_passing_math                
```

```python
#create variable that filters for passing reading scores - greater than 70
passing_reading = school_complete.loc[(school_complete['reading_score'] >= 70),:]
percent_passing_reading = (passing_reading['Student ID'].count() / total_students) *100
percent_passing_reading
```

```python
#create variable that filters for those who passed both math and reading
passing_both = school_complete.loc[(school_complete['reading_score'] >= 70) &
                                   (school_complete['math_score'] >= 70)
                                   ,:]
percent_passing_both = (passing_both['Student ID'].count() / total_students) *100
percent_passing_both
```

```python
#create a summary dataframe for the previous calculations
district_summary = pd.DataFrame({'Total Schools': [school_count],'Total Students': [total_students],'Total Budget': [total_budget],'Average Math Score':[average_math_score],'Average Reading Score': [average_reading_score],
                                '% Passing Math': [percent_passing_math],'% Passing Reading':[percent_passing_reading],'% Passing Overall':[percent_passing_both]})
district_summary


```

## School Summary

```python
#select the school type
school_types = school_data.set_index(["school_name"])["type"]
school_types.head()
```

```python
#find how many students attend each school
students_per_school = school_complete['school_name'].value_counts()
students_per_school.head()
```

```python
#find the per school budget
per_school_budget = school_complete.groupby(['school_name']).mean()['budget']
per_school_budget
```

```python
#find the budget per student
per_school_capita = per_school_budget / students_per_school
per_school_capita
```

```python
#find the average math schore by school
per_school_math = school_complete.groupby(['school_name']).mean()['math_score']
per_school_math
```

```python
#find the average reading score by school
per_school_reading = school_complete.groupby(['school_name']).mean()['reading_score']
per_school_reading
```

```python
#create a list of students with passing scores
passing_math_list = school_complete.loc[(school_complete['math_score'] >= 70),:]
passing_reading_list = school_complete.loc[(school_complete['reading_score'] >= 70),:]
passing_both_list = school_complete.loc[(school_complete['math_score'] >= 70) &
                                         (school_complete['reading_score'] >= 70) ,:]
```

```python
#find the percent of students who passed by diving the above by total students at each school
percent_passing_math = passing_math_list.groupby('school_name').count()['student_name'] / students_per_school *100
percent_passing_reading = passing_reading_list.groupby('school_name').count()['student_name'] / students_per_school *100
percent_passing_both = passing_both_list.groupby('school_name').count()['student_name'] / students_per_school *100
percent_passing_both
```

```python
#create a summary dataframe of the above calculations and change the formats of the spend columns
per_school_summary = pd.DataFrame({'School Type':school_types,'Total Students':students_per_school,'Total School Budget':per_school_budget,'Budget per Student':per_school_capita,'Average Math Score':per_school_math,'Average Reading Score':per_school_reading,'Percent Passing Math':percent_passing_math,'Percent Passing Reading':percent_passing_reading,'Percent Passing Both':percent_passing_both})
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Budget per Student"] = per_school_summary["Budget per Student"].map("${:,.2f}".format)
per_school_summary 
```

## Highest-Performing Schools (by % Overall Passing)

```python
#sort the top 5 schools based on percent who passed both math and reading
top_schools = per_school_summary.sort_values(['Percent Passing Both'],ascending=False)
top_schools.head(5)
```

## Lowest-Performing Schools (by % Overall Passing)

```python
#sort the bottom 5 schools based on percent who passed both
bottom_schools = per_school_summary.sort_values(['Percent Passing Both'],ascending=True)
bottom_schools.head(5)
```

## Math Scores by Grade

```python
#scores by grade
ninth_grade = school_complete.loc[(school_complete['grade'] == '9th')]
tenth_grade = school_complete.loc[(school_complete['grade'] == '10th')]
eleventh_grade = school_complete.loc[(school_complete['grade'] == '11th')]
twelfth_grade = school_complete.loc[(school_complete['grade'] == '12th')]
```

```python
#group the mean math scores for each grade by school
ninth_graders_math = ninth_grade.groupby(['school_name']).mean()['math_score']
tenth_graders_math = tenth_grade.groupby(['school_name']).mean()['math_score']
eleventh_graders_math = eleventh_grade.groupby(['school_name']).mean()['math_score']
twelfth_graders_math = twelfth_grade.groupby(['school_name']).mean()['math_score']
ninth_graders_math
```

```python
#create a dataframe that summarize the mean math scores by grade
math_scores_by_grade = pd.DataFrame({'9th grade':ninth_graders_math,'10th grade':tenth_graders_math,'11th grade': eleventh_graders_math, '12th grade':twelfth_graders_math})
math_scores_by_grade.index.name = None
math_scores_by_grade
```

## Reading Scores by Grade

```python
#group the mean reading scores for each grade by school
ninth_graders_reading = ninth_grade.groupby(['school_name']).mean()['reading_score']
tenth_graders_reading = tenth_grade.groupby(['school_name']).mean()['reading_score']
eleventh_graders_reading = eleventh_grade.groupby(['school_name']).mean()['reading_score']
twelfth_graders_reading = twelfth_grade.groupby(['school_name']).mean()['reading_score']
```

```python
#create a dataframe that summarizes the mean reading scores by grade
reading_scores_by_grade = pd.DataFrame({'9th grade':ninth_graders_reading,'10th grade':tenth_graders_reading,'11th grade': eleventh_graders_reading, '12th grade':twelfth_graders_reading})
reading_scores_by_grade.index.name = None
reading_scores_by_grade
```

## Scores by School Spend

```python
#create bins and labels based on per student spend
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#convert budget per student back to flot
per_school_summary['Budget per Student']=per_school_summary['Budget per Student'].str.replace('$','',regex=True)
per_school_summary['Budget per Student']=per_school_summary['Budget per Student'].astype('float64')
per_school_summary
```

```python
#put the budget per student into the assigned bins
summary_by_spend = pd.cut(per_school_summary['Budget per Student'],spending_bins,labels=labels)
per_school_summary['Spending Ranges per Student']=summary_by_spend
per_school_summary
```

```python
#group mean scores and percent passing by the new bins
spending_math_scores = per_school_summary.groupby(["Spending Ranges per Student"]).mean()["Average Math Score"]
spending_reading_scores = per_school_summary.groupby(["Spending Ranges per Student"]).mean()["Average Reading Score"]
spending_passing_math = per_school_summary.groupby(["Spending Ranges per Student"]).mean()["Percent Passing Math"]
spending_passing_reading = per_school_summary.groupby(["Spending Ranges per Student"]).mean()["Percent Passing Reading"]
overall_passing_spending = per_school_summary.groupby(["Spending Ranges per Student"]).mean()["Percent Passing Both"]
```

```python
#create dataframe to show average scores by bins
spending_summary = pd.DataFrame({'Average Math Score':spending_math_scores,'Average Reading Score':spending_reading_scores,'% Passing Math':spending_passing_math,'% Passing Reading':spending_passing_reading,'% Passing Both':overall_passing_spending})
spending_summary
```

## Scores by School Size

```python
#create bins and labels for school size
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#put total students into the assigned bins
by_total_students = pd.cut(per_school_summary['Total Students'],size_bins,labels=labels)
per_school_summary['Student Size Ranges'] = by_total_students
per_school_summary
```

```python
#group mean scores and percent passing by the size bins
size_math_scores = per_school_summary.groupby(["Student Size Ranges"]).mean()["Average Math Score"]
size_reading_scores = per_school_summary.groupby(["Student Size Ranges"]).mean()["Average Reading Score"]
size_passing_math = per_school_summary.groupby(["Student Size Ranges"]).mean()["Percent Passing Math"]
size_passing_reading = per_school_summary.groupby(["Student Size Ranges"]).mean()["Percent Passing Reading"]
overall_passing_spending_size = per_school_summary.groupby(["Student Size Ranges"]).mean()["Percent Passing Both"]
```

```python
#create dataframe to show mean scores and percent passing by school size
size_summary = pd.DataFrame({'Average Math Score':size_math_scores,'Average Reading Score':size_reading_scores,'% Passing Math':size_passing_math,'% Passing Reading':size_passing_reading,'% Passing Both':overall_passing_spending_size})
size_summary
```

## Scores by School Type

```python
#group mean scores and percent passing by school type
type_math_scores = per_school_summary.groupby(["School Type"]).mean()["Average Math Score"]
type_reading_scores = per_school_summary.groupby(["School Type"]).mean()["Average Reading Score"]
type_passing_math = per_school_summary.groupby(["School Type"]).mean()["Percent Passing Math"]
type_passing_reading = per_school_summary.groupby(["School Type"]).mean()["Percent Passing Reading"]
overall_passing_spending_type = per_school_summary.groupby(["School Type"]).mean()["Percent Passing Both"]
```

```python
#create new dataframe showing scores by school type
type_summary = pd.DataFrame({'Average Math Score':type_math_scores,'Average Reading Score':type_reading_scores,'% Passing Math':type_passing_math,'% Passing Reading':type_passing_reading,'% Passing Both':overall_passing_spending_type})
type_summary
```
